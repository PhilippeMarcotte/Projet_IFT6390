## III) Preprocessing des données


```python
from sklearn.preprocessing import StandardScaler
print('\nPreprocessing on Training and Validation Data...')
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import itertools
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

class svm_model:
    
    def __init__(self, C_value=1, kernel_value='rbf'):
        self.C_value = C_value
        self.kernel_value = kernel_value
        self.classifier = SVC(C = self.C_value, kernel = self.kernel_value, random_state=0)
        print('\n====== SVM Classifier with C = ' + str(self.C_value) + ' kernel = ' + self.kernel_value + '=========')
        
    def svm_train(self, X_train, y_train):
        self.classifier.fit(X_train, y_train)
        
    def svm_predict(self, X_test):
        self.y_pred = self.classifier.predict(X_test)
        
    def svm_print_accuracy_score(self, y_test):
        print('\nCalculating Accuracy of Predictions...')
        self.y_test = y_test
        print("Accuracy score : ", accuracy_score(self.y_test, self.y_pred))

    def search_by_gridSearch_with_only_SVM(self, X_train, y_train, C_values, cv=10): 
        hyerparams = [{'C': C_values, 'kernel': ['linear']},
                      {'C': C_values, 'kernel': ['rbf']}]
        self.grid_search = GridSearchCV(SVC(),
                                      param_grid = hyerparams,
                                      scoring = 'accuracy',
                                      cv = cv,
                                      n_jobs = -1)
        self.grid_search.fit(X_train, y_train)
      
    def print_best_score_find_by_gridSearch(self):
        print("Meilleur résultat de la validation croisée :", self.grid_search.best_score_)
        print("Meilleur(s) hyperparamètre(s) sur le jeu d'entraînement:", self.grid_search.best_params_)
      
    def print_score_of_generalisation_on_test_set(self, X_test, y_test):
        # Après avoir trouvé la meilleure combinaison de paramètres, il est 
        # important de calculer l'erreur de généralisation sur le test_set
        self.grid_search_y_pred = grid_search.predict(X_test)
        print("\nResultat sur le jeu de test : %0.3f" % metrics.accuracy_score(y_test, self.grid_search_y_pred))

    def plot_confusion_matrix(self): 
        # Plot Confusion Matrix for Test Data
        normalize=False
        cm = confusion_matrix(self.y_test, self.y_pred)
        cmap=plt.cm.Blues
        classes = ['T-Shirt/Top','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle Boot']
        
        plt.imshow(cm, interpolation='nearest', cmap=cmap)
        plt.title('Confusion Matrix for Test Data')
        plt.colorbar()
        tick_marks = np.arange(len(classes))
        plt.xticks(tick_marks, classes, rotation=45)
        plt.yticks(tick_marks, classes)
        
        fmt = '.2f' if normalize else 'd'
        thresh = cm.max() / 2.
        for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
            plt.text(j, i, format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")

        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.tight_layout()
        plt.grid('off')
        plt.show()
```
